```python
import random

trials = 10000
heads = 0
tails = 0

for _ in range(trials):
    toss = random.choice(['H', 'T'])
    if toss == 'H':
        heads += 1
    else:
        tails += 1

print("Heads Probability:", heads / trials)
print("Tails Probability:", tails / trials)
```

```python
import random

trials = 10000
count_sum_7 = 0

for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    if die1 + die2 == 7:
        count_sum_7 += 1

print("Probability of sum 7:", count_sum_7 / trials)
```

```python
import random

def probability_at_least_one_six(trials=10000):
    success = 0

    for _ in range(trials):
        found_six = False
        for _ in range(10):
            if random.randint(1, 6) == 6:
                found_six = True
                break
        if found_six:
            success += 1

    return success / trials

print("Estimated Probability:", probability_at_least_one_six())
```

```python
import random

colors = ['R'] * 5 + ['G'] * 7 + ['B'] * 8
trials = 1000

previous = None
blue_prev = 0
blue_then_red = 0

for _ in range(trials):
    current = random.choice(colors)

    if previous == 'B':
        blue_prev += 1
        if current == 'R':
            blue_then_red += 1

    previous = current

prob_red_given_blue = blue_then_red / blue_prev
print("P(Red | Previous Blue):", prob_red_given_blue)
```

```python
import numpy as np

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

sample = np.random.choice(values, size=1000, p=probabilities)

mean = np.mean(sample)
variance = np.var(sample)
std_dev = np.std(sample)

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)

```

```python
import numpy as np
import matplotlib.pyplot as plt

population = np.random.uniform(0, 1, 10000)

sample_means = []
for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

plt.hist(population, bins=30, density=True, alpha=0.6)
plt.title("Uniform Distribution")
plt.show()

plt.hist(sample_means, bins=30, density=True, alpha=0.6)
plt.title("Distribution of Sample Means (CLT)")
plt.show()
```
